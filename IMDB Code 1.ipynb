```r
# Please use selector gadget to get the tags:
# https://chrome.google.com/webstore/detail/selectorgadget/mhjhnkcfbdhnjickkkdbjoemdmbfginb?hl=en

# A good article on rvest, check this link:
# https://analyticsvidhya.com/blog/2017/03/beginners-guide-on-web-scraping-in-r-using-rvest-with-hands-on-knowledge/

# install.packages("rvest")
# install.packages("xlsx")
# install.packages("stringr")
library("rvest") # To web-scrap
library("stringr") # For post-processing of scraped data
library("xlsx") #To write the scraped data to an excel file
```

```r
# Load the URL
url <- 'https://www.imdb.com/chart/top'
```

```r
#Read the content
webpage <- read_html(url)
```

```r
#Scrap the desired data
scraped_data_html <- html_nodes(webpage,'.titleColumn')
```

```r
#convert to HTML
scraped_data_text <- html_text(scraped_data_html)
scraped_data_text
```

```r
#post processing

#1. Remove '\n'
scraped_data_text <- str_replace_all(scraped_data_text, '\n      ', '')
scraped_data_text

# Write to excel file
write.xlsx(scraped_data_text, 'top100_list.xlsx')
```
